* Model Function

洧녭洧녻,洧녪(洧논(洧녰))=洧녻洧논(洧녰)+洧녪

```python
w = 200
b = 100

#if we use w = 100, and b=100, it will different value, so, 200 and 100 suits the data well
print(f"w:{w}")
print(f"b:{b}")
```

洧녭洧녻,洧녪(洧논(洧녰))
  for your two data points. You can explicitly write this out for each data point as -

for 洧논(0)
, f_wb = w * x[0] + b

for 洧논(1)
, f_wb = w * x[1] + b

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    #taking the number of samples
    f_wb = np.zeros(m) #initializing a new array to 0 of length m
    for i in range(m):
        f_wb[i] = w*x[i]+b
    
    return f_wb
```

Call the *compute_model_output* function and plot the output

```python
tmp_f_wb = compute_model_output(x_train, w, b,)
print(f"The model function is {tmp_f_wb}")
# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
